```python
import pdf2image, os, pytesseract
from img2table.ocr import TesseractOCR
from img2table.document import PDF
from datetime import datetime
import cv2
from PIL import Image as PILImage
from IPython.display import display_html
from tabula import read_pdf
from tabulate import tabulate
```

```python
output_text = ''
for pdf_page in pdf2image.convert_from_path("iep.pdf"):
    output_text += pytesseract.image_to_string(pdf_page)
#print(output_text)
```

```python
output_text = ''
pdf = PDF(src="iep.pdf")
extracted_tables = pdf.extract_tables(ocr=TesseractOCR(n_threads=1, lang='eng'),
                                                implicit_rows=False,
                                                borderless_tables=False,
                                                min_confidence=80)
```

```python
for page, tables in extracted_tables.items():
    for idx, table in enumerate(tables):
        display_html(table.html_repr(title=f"Page {page + 1} - Extracted table No{idx + 1}"), raw=True)
```
